### 1. Uploading data

```stata
use "https://github.com/worldbank/Stata-IE-Visual-Library/raw/develop/Library/Density%20plots/Density%20plot%20with%20averages/data.dta", clear
```

### 2. Calculate the means

```stata
qui sum 	revenue if post == 0
local 	pre_mean = r(mean)
qui sum		revenue if post == 1
local 	post_mean = r(mean)
```

### 3. Creating the graph

```stata
twoway 	(kdensity revenue if post == 0, color(gs10)) ///
		(kdensity revenue if post == 1, color(emerald)), ///
		xline(`pre_mean', lcolor(gs12) lpattern(dash)) ///
		xline(`post_mean', lcolor(eltgreen) lpattern(dash)) ///
		legend(order(1 "Pre-treatment" 2 "Post-treatment")) ///
		xtitle(Agriculture revenue (BRL thousands)) ///
		ytitle(Density) ///
		bgcolor (white) graphregion(color(white))
```

### 4. Exporting the graph

```stata
qui gr export "figure.png", width(5000) replace
```
